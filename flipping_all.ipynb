```python
import os
import cv2

# Replace this with the path to your 'videos' folder
VIDEOS_FOLDER = "videos-Copy1"

# Loop through all files in the folder
for video_file in os.listdir(VIDEOS_FOLDER):
    video_path = os.path.join(VIDEOS_FOLDER, video_file)
    
    # Skip if it's not a video
    if not os.path.isfile(video_path) or not video_file.lower().endswith((".mp4", ".avi", ".mov", ".mkv")):
        continue
    
    print(f"Flipping {video_file} ...")
    
    # Open video
    cap = cv2.VideoCapture(video_path)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    # Save flipped video as a new file
    flipped_path = os.path.join(VIDEOS_FOLDER, os.path.splitext(video_file)[0] + "_flipped.mp4")
    out = cv2.VideoWriter(flipped_path, fourcc, fps, (width, height))
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        flipped = cv2.flip(frame, 1)  # Horizontal flip
        out.write(flipped)
    
    cap.release()
    out.release()

print("All videos have been flipped and saved with '_flipped' suffix.")
```
